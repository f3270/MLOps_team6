# Dataset Information

The data set describes pediatric patients with several hematologic diseases: malignant disorders (i.a. acute lymphoblastic leukemia, acute myelogenous leukemia, chronic myelogenous leukemia, myelodysplastic syndrome) and nonmalignant cases (i.a. severe aplastic anemia, Fanconi anemia, with X-linked adrenoleukodystrophy). All patients were subject to the unmanipulated allogeneic unrelated donor hematopoietic stem cell transplantation. 
 
The motivation of the study was to identify the most important factors influencing the success or failure of the transplantation procedure. In particular, the aim was to verify the hypothesis that increased dosage of CD34+ cells / kg extends overall survival time without simultaneous occurrence of undesirable events affecting patients' quality of life (KawÅ‚ak et al., 2010).

The data set has been used in our work concerning survival rules (WrÃ³bel et al., 2017) and user-guided rule induction (Sikora et al., 2019). The authors of the research on stem cell transplantation (KawÅ‚ak et al., 2010) who inspired our study also contributed to the set.

# Descripción del Dataset

## Bmt Dataset

El conjunto de datos describe pacientes pediátricos con varias enfermedades hematológicas:

- **Trastornos malignos**: pacientes con leucemia linfoblástica aguda, leucemia mieloide aguda, leucemia mieloide crónica, síndrome mielodisplásico.
- **Casos no malignos**: pacientes con anemia aplásica severa, anemia de Fanconi, adrenoleucodistrofia ligada al cromosoma X.

Todos los pacientes fueron sometidos a un trasplante de células madre hematopoyéticas alogénico no manipulado de un donante no relacionado. El objetivo del estudio fue identificar los factores más importantes que influyen en el éxito o fracaso del procedimiento de trasplante. En particular, se verificó la hipótesis de que un aumento en la dosis de células CD34+/kg extiende el tiempo de supervivencia general sin la aparición de eventos indeseables que afecten la calidad de vida de los pacientes.

### Información del dataset

El conjunto contiene 187 ejemplos caracterizados por 37 atributos:

- **Recipientgender**: Género del receptor (Hombre - 1, Mujer - 0).
- **Stemcellsource**: Fuente de células madre hematopoyéticas (Sangre periférica - 1, Médula ósea - 0).
- **Donorage**: Edad del donante en el momento de la aféresis de células madre hematopoyéticas.
- **Donorage35**: Edad del donante <35 - 0, Edad del donante >=35 - 1.
- **IIIV**: Desarrollo de enfermedad injerto contra huésped aguda en estadio II, III o IV (Sí - 1, No - 0).
- **Gendermatch**: Compatibilidad de género entre el donante y receptor (Femenino a Masculino - 1, Otro - 0).
- **DonorABO**: Grupo sanguíneo ABO del donante (0 - 0, A - 1, B - (-1), AB - 2).
- **RecipientABO**: Grupo sanguíneo ABO del receptor (0 - 0, A - 1, B - (-1), AB - 2).
- **RecipientRh**: Presencia del factor Rh en los glóbulos rojos del receptor ('+' - 1, '-' - 0).
- **ABOMatch**: Compatibilidad ABO entre donante y receptor (Compatible - 1, No compatible - 1).
- **CMVstatus**: Compatibilidad serológica de citomegalovirus entre donante y receptor (mayor valor, menor compatibilidad).
- **RecipientCMV**: Presencia de infección por citomegalovirus en el receptor antes del trasplante (Presente - 1, Ausente - 0).
- **Disease**: Tipo de enfermedad (ALL, AML, crónico, no maligno, linfoma).
- **Riskgroup**: Grupo de riesgo (Alto - 1, Bajo - 0).
- **Txpostrelapse**: Segundo trasplante de médula ósea después de la recaída (No - 0; Sí - 1).
- **Diseasegroup**: Tipo de enfermedad (Maligno - 1, No maligno - 0).
- **HLAmatch**: Compatibilidad del complejo mayor de histocompatibilidad (HLA) según criterios internacionales BFM SCT 2008 (10/10 - 0, 9/10 - 1, 8/10 - 2, 7/10 - 3).
- **HLAmismatch**: Compatibilidad HLA (Compatible - 0, No compatible - 1).
- **Antigen**: Diferencias en antígenos entre donante y receptor (sin diferencias - (-1), una diferencia - 0, dos o tres diferencias - 1 o 2).
- **Allel**: Diferencias en alelos entre donante y receptor (sin diferencias - (-1), una diferencia - 0, dos, tres o cuatro diferencias - 1, 2, 3).
- **HLAgrI**: Tipo de diferencia entre donante y receptor en HLA (compatible - 0, diferencia en un antígeno - 1, diferencia en un alelo - 2, diferencia en célula DRB1 - 3, dos diferencias en alelos o antígenos - 4 o 5).
- **Recipientage**: Edad del receptor en el momento del trasplante.
- **Recipientage10**: Edad del receptor <10 - 0, Edad del receptor >=10 - 1.
- **Recipientageint**: Edad del receptor en intervalos (0,5] - 0, (5,10] - 1, (10,20] - 2.
- **Relapse**: Reaparición de la enfermedad (No - 0, Sí - 1).
- **aGvHDIIIIV**: Desarrollo de enfermedad injerto contra huésped aguda en estadio III o IV (Sí - 0, No - 1).
- **extcGvHD**: Desarrollo de enfermedad injerto contra huésped crónica extensa (Sí - 0, No - 1).
- **CD34kgx10d6**: Dosis de células CD34+ por kg de peso corporal del receptor (10^6/kg).
- **CD3dCD34**: Relación entre células CD3+ y CD34+.
- **CD3dkgx10d8**: Dosis de células CD3+ por kg de peso corporal del receptor (10^8/kg).
- **Rbodymass**: Masa corporal del receptor en el momento del trasplante.
- **ANCrecovery**: Tiempo hasta la recuperación de neutrófilos (cuenta de neutrófilos >0.5 x 10^9/L).
- **PLTrecovery**: Tiempo hasta la recuperación de plaquetas (cuenta de plaquetas >50,000/mm3).
- **time_to_aGvHD_III_IV**: Tiempo hasta el desarrollo de enfermedad injerto contra huésped aguda en estadio III o IV.
- **survival_time**: Tiempo de supervivencia (numérico).
- **survival_status**: Estado de supervivencia.
